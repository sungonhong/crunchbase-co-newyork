## Question 1: Read the CSV file into a Pandas data frame and drop records where company names are None values:
- Print the count of records (rows) in the resulting data frame
- Print the first 5 records of the data frame

```python
import pandas as pd 

df = pd.read_csv("./data/crunchbase_odm_orgs.csv") 

print("Columns:", df.columns)
print(df['type'].unique())
print(df['primary_role'].unique())

print(f"Total {len(df)} records")
print("=== dropping records where company names are None values ===") 


#Condition to drop is with Primary_role == 'company' and 'Name' being Null 
to_drop = ((df['primary_role']=='company') & (df['name'].isna()))
df_clean = df[~to_drop] 
print(f"Total {len(df_clean)} records")

print("First 5:") 
print(df_clean.head())

```

## Question 2: Create a data frame that contains only the records of USA-based companies whose name starts with "Ac":
- Print the count of records (rows) in the resulting data frame
- Print the first 5 records of the data frame


```python

print(f"Country codes are: {df_clean['country_code'].unique()}")

filter1 = df_clean['country_code'] == 'USA' 
filter2 = df_clean['name'].str.match(r'^Ac',case=False) 
df_gp = df_clean[filter1 & filter2]

print(f"Desired data: {len(df_gp)} records")

print(df_gp.head())
```

## Question 3: Convert the data frame from the previous step into a list of JSON objects:
- Print the count of JSON objects
- Print the first 5 JSON objects
- Write all JSON objects into a text (JSON string) file
- Print the number of records in the resulting file


```python
import json 

#convert to dictionaries 
json_objects = df_clean.to_dict(orient='records')
print(f"{len(json_objects)} JSON objects") 
for obj in json_objects[:5]: 
    print(obj)

json_filename = './aaagroup.json'
#dump to string and write to a file 
with open(json_filename, 'w') as f:
    f.write(json.dumps(json_objects))

#let's read the file 
loaded_json = None 
with open(json_filename, "r") as f:
    loaded_json = json.load(f)

if loaded_json: 
    print(f"\n{len(loaded_json)} records in JSON file") 
else: 
    print("\nCouldn't read the JSON file or no records") 
```

## Question 5: Implement Flask API functions to Return a list of Crunchbase companies based in New York
- Take a user input through an API endpoint to return a list of JSON objects from Webhose file where the "title" fields contain the query string 

```python
import sqlite3

#table_name is the name of the table that will be created in sqlite3 db 
table_name = "company" 

############ ############ ############ 
#creates the db, connect and run SQL 
#query is the SQL commands given by the caller 
######################### ############ 
def run_query(query=[]): 

    if len(query) == 0: 
        raise KeyError 

    retVal = []  
    conn = None 
    try: 
        conn = sqlite3.connect('./data/newyork_co.db')
    
        #create a sql db using the dataframe data 
        df_newyork_co.to_sql(table_name, conn, if_exists="replace", index=False)

        cursor = conn.cursor()
        cursor.execute(query)
        retVal.append( [description[0] for description in cursor.description]) 
        #column names 
        retVal.append(cursor.fetchall()) 
        return retVal 

    except Exception as e: 
        raise e 
        
    finally:
        if conn: 
            conn.close()
        
```

```python
#pip install flask 
```

```python
from flask import request

@app.route('/query', methods=['GET', 'POST'])
def query(): 
    if request.method == 'GET':
        # Allow users to put in their query as title 
        return '''<form method="POST">
                    <p> The table name is company</p>
                    <label for="title">Enter your query:</label>
                    <input type="text" name="title" id="title" placeholder="Enter your query" size=100> 
                    <button type="submit">Submit</button>
                  </form>'''
    else:
        # Process the submitted query 
        try:
            title = request.form['title']
            
            if title is None or len(title) == 0: 
                return "Query not submitted. Please enter your query again"

            elif title.casefold().startswith('SELECT'.casefold()) == False:  
                return "Only Select SQL query accepted. The first command is expected to be SELECT"
                
            else: 
                data = run_query(title)

                if data is None or len(data) == 0: 
                    return "No data returned." 
                else: 
                    columns = data[0]
                    list_of_dicts = [dict(zip(columns, row)) for row in data[1]]
                    return render_template('index2.html',  records=list_of_dicts)   
                
        except KeyError:
            # Handle the case where the 'username' key is not present in the form data
            return "Error processing query. Please enter your query again"

        except Exception as e: 
            return f"Error processing query. {e}"
```

```python
app.run(host='localhost', port=5004, use_reloader = False, debug=True) 
```
